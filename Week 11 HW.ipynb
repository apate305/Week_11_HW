ARIMA is a statistical model used for analysing time series data:


Autoregression (AR) is where past values including the errors is used to predict future values. The assumption is that the past and future data will have some correlation

In other words, the WAY it changes can change (so its not completely stationary).

How do we get order of integration? subtract an observation from the observation at the previous time stamp.

Moving Average (MA) -helps smooth out data over a particular time period by creating a consantly averaged value. The impacts of random short-term fluctuations are decreased.

How do we get MA: add up all values in a given time frame and divide by the number of values.

6.	For two of the cities you found in question 5 (there are many possible answers), clean up the data via rolling averages and plot a clear visual. Explain how rolling average differs from moving average (MA) model.

Rolling and moving average helps smooth out data over a particular time period by creating a consantly averaged value. So impacts of random short term fluctuations are decreased. 


Rolling average is a type of moving average. It gives the average for all the data up untill the point specified. For example, the rolling average of profit in April would be calculated by adding the profits in January, February March and April and then dividing the sum by 4.


On the other hand, moving average is the average of a data set for a chosen period of time. For example, the moving average of profits in April with a period of 2 would be adding numbers in March and April and then dividing by two.
